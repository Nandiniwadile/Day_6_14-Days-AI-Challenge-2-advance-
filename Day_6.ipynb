```python
from pyspark.sql import SparkSession

# Sample data
data = [
    (10, 1, 500.0, 500.0, 1),
    (5, 0, 0.0, 0.0, 0),
    (20, 3, 1500.0, 500.0, 1),
    (8, 0, 0.0, 0.0, 0),
    (15, 2, 800.0, 400.0, 1),
    (7, 0, 0.0, 0.0, 0),
    (25, 5, 3000.0, 600.0, 1),
    (4, 0, 0.0, 0.0, 0),
    (12, 1, 400.0, 400.0, 1),
    (6, 0, 0.0, 0.0, 0)
]

columns = ["total_events","purchases","total_spent","avg_price","purchased"]

training_data = spark.createDataFrame(data, columns)

training_data.show()
```

```python
from pyspark.ml.feature import VectorAssembler

assembler = VectorAssembler(
    inputCols=["total_events","purchases","total_spent","avg_price"],
    outputCol="features"
)

data = assembler.transform(training_data).select("features","purchased")
```

```python
train_data, test_data = data.randomSplit([0.8, 0.2], seed=42)
```
